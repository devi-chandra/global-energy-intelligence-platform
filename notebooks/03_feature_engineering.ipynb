```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/clean_energy_data.csv")

print(df.shape)
```

```python
candidate_features = [
    "year",
    "population",
    "gdp",
    "electricity_generation",
    "electricity_demand",
    "coal_consumption",
    "gas_consumption",
    "oil_consumption",
    "renewables_consumption",
    "low_carbon_consumption",
    "greenhouse_gas_emissions",
    "primary_energy_consumption"
]

df[candidate_features].isnull().sum().sort_values(ascending=False)
```

```python
df["energy_per_gdp_calc"] = (
    df["primary_energy_consumption"] /
    df["gdp"]
)
```

```python
df["renewable_ratio"] = (
    df["renewables_consumption"] /
    df["primary_energy_consumption"]
)
```

```python
df["fossil_ratio"] = (
    df["fossil_fuel_consumption"] /
    df["primary_energy_consumption"]
)
```

```python
df["electricity_per_person"] = (
    df["electricity_generation"] /
    df["population"]
)
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
df[[
    "energy_per_gdp_calc",
    "renewable_ratio",
    "fossil_ratio",
    "electricity_per_person"
]].isnull().sum()
```

```python
df.to_csv(
    "../data/engineered_energy_data.csv",
    index=False
)
```
