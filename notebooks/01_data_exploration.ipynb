```python
# Import the libraries we need
import pandas as pd
import numpy as np

# Load the dataset into a "DataFrame" (a table)
df = pd.read_csv("../data/owid-energy-data.csv")

# Show the first 5 rows
df.head()
```

```python
# Show the shape: (number of rows, number of columns)
df.shape
```

```python
# List every column name
df.columns.tolist()
```

```python
# Get a technical summary of the data
df.info()
```

```python
# Define the columns we actually need for this project
key_columns = [
    "country",
    "year",
    "iso_code",
    "population",
    "gdp",
    "primary_energy_consumption",
    "renewables_share_energy"
]

# Create a new DataFrame with ONLY these columns
df_energy = df[key_columns]

# Look at the first 5 rows of our focused data
df_energy.head()
```

```python
# Technical summary of our focused dataset
df_energy.info()
```

```python
# Count how many missing (NaN) values are in each column
df_energy.isnull().sum()
```

```python
# Show missing values as a percentage of total rows
(df_energy.isnull().sum() / len(df_energy) * 100).round(2)
```

```python
# Make a copy of df_energy to clean (keep original safe)
df_clean = df_energy.copy()

# Check how many rows BEFORE cleaning
print("Rows before removing regions:", len(df_clean))

# Remove rows where iso_code is missing (these are regions like ASEAN, World, etc.)
df_clean = df_clean.dropna(subset=["iso_code"])

# Check how many rows AFTER
print("Rows after removing regions:", len(df_clean))
```

```python
# Check rows before filtering years
print("Rows before year filter:", len(df_clean))

# Keep only rows from 1990 onwards
df_clean = df_clean[df_clean["year"] >= 1990]

# Check rows after filtering
print("Rows after year filter:", len(df_clean))

# Check the year range now
print("Year range:", df_clean["year"].min(), "to", df_clean["year"].max())
```

```python
# Rows before dropping
print("Rows before dropping missing critical values:", len(df_clean))

# Drop rows where ANY of these 3 important columns is missing
df_clean = df_clean.dropna(subset=[
    "primary_energy_consumption",
    "population",
    "gdp"
])

# Rows after dropping
print("Rows after dropping missing critical values:", len(df_clean))
```

```python
# See remaining missing values per column
df_clean.isnull().sum()
```
